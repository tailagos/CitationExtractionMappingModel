```python
# LLM Citation Mapping Pipeline

# ===============================
# 1. Environment Setup
# ===============================
!pip install datasets faiss-cpu sentence-transformers openai

import os
import faiss
import torch
import numpy as np
from datasets import load_dataset
from sentence_transformers import SentenceTransformer

# ===============================
# 2. Dataset Loading & Preprocessing
# ===============================
squad = load_dataset("squad")
contexts = [item['context'] for item in squad['train']]
questions = [item['question'] for item in squad['train']]

# ===============================
# 3. Sentence Embedding & FAISS Indexing
# ===============================
model = SentenceTransformer('all-MiniLM-L6-v2')
context_embeddings = model.encode(contexts, convert_to_tensor=False, show_progress_bar=True)

index = faiss.IndexFlatL2(len(context_embeddings[0]))
index.add(np.array(context_embeddings))
```

```python
# ===============================
# 4. Semantic Retrieval Function
# ===============================
def retrieve_passages(query, top_k=5):
    query_embedding = model.encode([query])
    distances, indices = index.search(np.array(query_embedding), top_k)
    return [contexts[i] for i in indices[0]]
```

```python
# ===============================
# 5. Prompt Engineering for GPT-3.5
# ===============================
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-base")
model_gen = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-base")

def generate_answer_with_citations(question, passages):
    joined_context = "\n".join([f"[Source {i+1}] {p}" for i, p in enumerate(passages)]) # Use \n to create a newline
    prompt = f"Question: {question}\nSources: {joined_context}\nAnswer with citations:" # Use \n to create newlines
    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)
    outputs = model_gen.generate(**inputs, max_new_tokens=200)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
# ===============================
# 6. Citation Mapping (Embedding Similarity)
# ===============================
def map_citations(answer, passages):
    from sklearn.metrics.pairwise import cosine_similarity
    answer_segments = answer.split('.')  # crude segmentation
    passage_embeddings = model.encode(passages)
    segment_embeddings = model.encode(answer_segments)

    mapping = []
    for seg, seg_emb in zip(answer_segments, segment_embeddings):
        sims = cosine_similarity([seg_emb], passage_embeddings)[0]
        best_match = np.argmax(sims)
        mapping.append((seg.strip(), f"[Source {best_match+1}]"))
    return mapping
```

```python
# ===============================
# 7. Evaluation Functions
# ===============================
def evaluate_precision_recall(pred_citations, true_citations):
    pred_set = set(pred_citations)
    true_set = set(true_citations)
    precision = len(pred_set & true_set) / len(pred_set) if pred_set else 0
    recall = len(pred_set & true_set) / len(true_set) if true_set else 0
    return precision, recall
```

```python
# ===============================
# 8. Example Run
# ===============================
question = questions[0]
passages = retrieve_passages(question)
answer = generate_answer_with_citations(question, passages)
mapping = map_citations(answer, passages)

print("\nGenerated Answer:\n", answer)
print("\nCitation Mapping:\n", mapping)
```
